# 01 — Strategic Asset Allocation (monthly)
This notebook pulls public ETF data and runs the SAA optimiser backtest.

```python
from wealth_lab.config import DEFAULT_UNIVERSE
from wealth_lab.data.prices_yf import fetch_adj_close, to_monthly_returns
from wealth_lab.saa.backtest import run_monthly_backtest

px = fetch_adj_close(list(DEFAULT_UNIVERSE.tickers), start='2010-01-01')
rets = to_monthly_returns(px).dropna()
bt = run_monthly_backtest(rets, lookback_months=60)
bt[['portfolio_nav']].tail()
```

```python
bt['portfolio_nav'].plot(title='SAA NAV')
```
